# Building with the Claude API

## Chat Bot Exercise

### Environment setup

```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

### Helper functions

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return message.content[0].text
```

### Chat Bot

Make a chat bot using the three helper functions we just put together

```python
# 1. Prompt the user to enter some input using the built-in "input" function
question = input("What burning question do you have today?")
messages = []
while question:
    # 2. Add it to a list of messages    
    add_user_message(messages, question)
    
    # 3. Call the API
    answer = chat(messages)
    
    # 4. Add generated text to the list of messages
    add_assistant_message(messages, answer)
    
    # 5. Print the generated text
    print("---")
    print(answer)
    print("---")
    
    # 6. Repeat from #1
    question = input("What other question do you have?")

"Have a great day!"
```
