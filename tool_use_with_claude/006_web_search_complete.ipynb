```python
# Load env variables and create client
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()

client = Anthropic()
model = "claude-sonnet-4-5"
```

```python
# Helper functions
from anthropic.types import Message


def add_user_message(messages, message):
    user_message = {
        "role": "user",
        "content": message.content if isinstance(message, Message) else message,
    }
    messages.append(user_message)


def add_assistant_message(messages, message):
    assistant_message = {
        "role": "assistant",
        "content": message.content if isinstance(message, Message) else message,
    }
    messages.append(assistant_message)


def chat(messages, system=None, temperature=1.0, stop_sequences=[], tools=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
        "stop_sequences": stop_sequences,
    }

    if tools:
        params["tools"] = tools

    if system:
        params["system"] = system

    message = client.messages.create(**params)
    return message


def text_from_message(message):
    return "\n".join([block.text for block in message.content if block.type == "text"])
```

```python
web_search_schema = {
    "type": "web_search_20250305",
    "name": "web_search",
    "max_uses": 5,
    "allowed_domains": ["nih.gov"],
}
```

```python
messages = []
add_user_message(
    messages,
    """
    What's the best exercise for gaining leg muscle?
    """,
)
response = chat(messages, tools=[web_search_schema])
response
```

The best exercises for gaining leg muscle are compound movements that target multiple muscle groups.

Here are the top options:

**1. Squats (Back Squats)**
- Often considered the king of leg exercises
- Targets quadriceps, hamstrings, glutes, and core
- Allows for heavy loading and progressive overload
- Variations include front squats, goblet squats, and Bulgarian split squats

**2. Deadlifts (Conventional or Romanian)**
- Excellent for posterior chain development (hamstrings, glutes, lower back)
- Romanian deadlifts specifically target hamstrings and glutes
- Great for overall leg and back strength

**3. Leg Press**
- Allows you to lift heavy weight safely
- Good for targeting quads with less technical demand than squats
- Easier on the lower back for those with limitations

**4. Lunges (Walking or Stationary)**
- Works each leg independently, helping correct imbalances
- Targets quads, glutes, and hamstrings
- Functional movement pattern

**5. Bulgarian Split Squats**
- Unilateral exercise for balanced development
- Excellent for glutes and quads
- Requires less weight than bilateral exercises

**Key Principles for Muscle Growth:**
- **Progressive overload**: Gradually increase weight, reps, or volume over time
- **Training frequency**: Train legs 2-3 times per week
- **Rep range**: 6-12 reps per set is ideal for hypertrophy
- **Proper nutrition**: Adequate protein (0.7-1g per pound of body weight) and caloric surplus
- **Rest and recovery**: Allow 48-72 hours between intense leg sessions

For most people, **squats and deadlifts form the foundation** of an effective leg-building program, supplemented with accessory exercises like lunges and leg press for complete development.
